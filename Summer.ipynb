# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
import numpy as np
df = pd.read_csv("summer.csv")
#df
```

### 1. In how many cities Summer Olympics is held so far?

```python
print("No. of cities in which olympic is held:")
len(df.City.unique())
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
df_gold = df[df['Medal']=='Gold']
df_gold['Sport'].value_counts().head().plot.bar(title ='Sport having most Number of Gold Medals So Far')
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
df['Sport'].value_counts().head().plot.bar(title ='Sport having  sport is having most number of medals so far')
```

### 4. Which player has won most number of medals? (Top 5)

```python
df['Athlete'].value_counts().head().plot.bar(title =' player has won most number of medals')
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
for index in range(len(df)):
  if(df.iloc[index]['Medal']=='Gold' and df.iloc[index]['Country']=='IND'):
    print("India won its first gold medal in summer olympics in:")
    print(df.iloc[index]['Year'])
    break
```
