# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
import numpy as np
df = pd.read_csv("summer.csv")
#df
```

### 1. In how many cities Summer Olympics is held so far?

```python
len(df.City.unique())
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
df_gold = df[df['Medal']=='Gold']
df_gold.groupby('Sport').count()['Medal'].sort_values(ascending=False).head().plot.bar(title ='Sport having most Number of Gold Medals So Far')
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
df.groupby('Sport').count()['Medal'].sort_values(ascending=False).head().plot.bar(title ='Sport having  sport is having most number of medals so far')
```

### 4. Which player has won most number of medals? (Top 5)

```python
df.groupby('Athlete').count()['Medal'].sort_values(ascending=False).head().plot.bar(title =' player has won most number of medals')
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
df_gold.groupby('Athlete').count()['Medal'].sort_values(ascending=False).head().plot.bar(title ='player has won most number Gold Medals of medals')
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
for index in range(len(df)):
  if(df.iloc[index]['Medal']=='Gold' and df.iloc[index]['Country']=='IND'):
    print("India won its first gold medal in summer olympics in:")
    print(df.iloc[index]['Year'])
    break
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
df.groupby('Event').count()['Athlete'].sort_values(ascending=False).head().plot.bar(title ='Sport having most Number of Gold Medals So Far')
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
df_women_gold=df_gold[df_gold['Gender']=='Women']
df_women_gold.groupby('Sport').count()['Medal'].sort_values(ascending=False).head().plot.bar(title ='Sport having most Number of Gold Medals So Far')
```
